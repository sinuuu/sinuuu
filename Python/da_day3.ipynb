```python
#[문제] Born과 Died 사이의 세월을 알아보려고 하나, 문자열이라 비교가 어려움
#datetime : 날짜형식의 포맷을 나타내는 수치열
born_datetime=pd.to_datetime(scientists['Born'],format='%Y-%m-%d')
#               to: ~형태로 자료를 변경해라!
born_datetime

died_datetime=pd.to_datetime(scientists['Died'],format='%Y-%m-%d')
died_datetime


# 포맷이 object -> datetime 으로 변경됨
# %Y-%m-%d : %Y(연도,4자리), %y(연도 2자리)
#시간 나타낼 때, 시는 h 분 M 초는 s
```

```python
scientists
```

```python
scientists['born_dt'],scientists['died_dt']=born_datetime,died_datetime
scientists
```

```python
# scientists
# scientists=['Born']-scientists['Died'] # string(문자형)간 연산은 불가하다
scientists['age_days_dt']=scientists['died_dt']-scientists['born_dt']  # 데이터 추가할 때
scientists
```

```python
#데이터의 순서를 바꾸고 싶다
import random
# random.seed(42)  # 렌덤넘버를 생성하기 위한 씨앗
random.shuffle(scientists['Age'])
print(scientists['Age'])
```
