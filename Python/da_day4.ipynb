등고선(contour) : 3차원 데이터를 시각화
- 입력변수가 2개, 출력변수 1개 
- 등고선 출력함수 : contour(등고선), contourf(색깔)
    

```python
def f(x,y):
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)
x=np.linspace(-3,3,256)
y=np.linspace(-3,3,256) #-3~3까지 256개의 숫자가 들어가 있는 형태
xx,yy=np.meshgrid(x,y) # mashgrid : 동일한 행렬을 생성해준다(x의 변수 길이가 256개 => 256줄로 그대로 생성해버림)

# 등고선을 만드려면 mashgrid 함수에 x를 넣어서 grid 행렬로 만들어주는게 선행이 되어야 함 -> 그 다음에 contour 함수 사용!
zz=f(xx,yy)  # x와 y의 값도 (256, 256) => 결과도 (256, 256)개가 나올 것임
zz.shape  


# xx.yy,zz 값 시각화 하기
plt.contourf(xx,yy,zz, cmap="jet",alpha=.75)
plt.contour(xx,yy,zz,colors='black')

print(xx.shape)
print("="*50)
print(yy)
```

```python
# x=[1,3]
# y=[2,4,6]

x=np.linspace(-10,10,101) # x값이 -10~10사이의 101개 출력
y=np.linspace(-10,10,101)

X,Y=np.meshgrid(x,y) # y요소가 2개이므로 x는 2행, x요소가 3개이므로 각 요소값이 3번 반복
# print(X)
# print("="*50)
# print(Y)

Z=np.sqrt(X**2+Y**2) #요소 간 곱셈(주의 행렬 곱셈 아님)
# #행렬의 곱셈 : [3,2][2,1]=>[3,1]
# print("="*50)
# print(Z)

cs=plt.contourf(X,Y,Z,100) #10: 레벨 / 넣으면 구분선이 10개로 나눠짐
plt.colorbar(cs)
# plt.contour(X,Y,Z,colors='black')
```
