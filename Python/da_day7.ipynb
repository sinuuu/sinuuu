```python
banks=pd.read_csv("data/banklist.csv")
banks.head() # Closing Date, Updated Date 날짜인데 문자열 형태임
# banks.info()
```

```python
# Closing Date, Updated Date 문자열 -> Datatime으로 바꿔주기

banks=pd.read_csv("data/banklist.csv",parse_dates=[5,6])
banks.head()
banks.info()
```

```python
# 언제 파산했는지 연도 찾기
# banks['Closing Date'].dt.year
# banks['Closing Date'].dt.quarter #  분기로 나눈 정보 추출

banks['Closing_quarter'],banks['Closing_year']=banks['Closing Date'].dt.quarter,banks['Closing Date'].dt.year
print(banks.head())
```

```python
# [문제] 연도별로 파산한 은행이 몇개인지 코드 작성
closing_year=banks.groupby(['Closing_year']).size()
print(closing_year)

# 연도별로 그룹화 한 다음 다시 분기별로 그룹화를 해서 출력
closing_year_q=banks.groupby(['Closing_year','Closing_quarter']).size()
print(closing_year_q)
```

```python
# 그래프로 출력하기

import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax=closing_year.plot()
plt.show


fig.ax=plt.subplots()
ax=closing_year_q.plot()
plt.show
```
