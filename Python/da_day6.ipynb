# error 인자에 설정할 수 있는 값
## ignore : 아무 작업도 하지 않음
## raise : 숫자로 변환할 수 없는 값이 있으면 오류 발생
## coerse : 숫자로 변환할 수 없는 값을 누락값으로 지정

```python
# ignore 사용

tips_sub_miss['total_bill']=pd.to_numeric(tips_sub_miss['total_bill'],errors='ignore')
# '오류들을 무시해라' 라는 의미 -> 여전히 total_bill은 문자열로!
# 숫자로 된 문자인 경우에는 to_numeric 사용시 숫자로 변환됨(숫자와 문자형태가 다 섞여있기 때문에 오류 메시지가 뜸)
tips_sub_miss.dtypes
```

```python
# coerse 사용

tips_sub_miss['total_bill']=pd.to_numeric(tips_sub_miss['total_bill'],errors='coerse')
# '강제로 누락값으로 바꿔라' -> total_bill은 float으로 바뀜

tips_sub_miss.dtypes
tips_sub_miss['total_bill'] # 1,3,5,7 행을 강제로 NaN으로 바꿔버림.. coerse 옵션
```

```python
# 전화번호를 패턴으로 저장하기

tel_num_spaces="123 456 7890"
m=re.match("\d{10}",tel_num_spaces)
print(m) # 공백으로 인해 매치가 되지 않음(None)

if m:
    print("매치됨")
else:
    print('매치안됨')
```

```python
df=pd.DataFrame({"a":[10,20,30],"b":[20,30,40]})
df

# a열의 값을 제곱하여 출력해보기
df['a']**2
```

```python
def my_sq(x):
    return x**2  # (제곱만)
def my_exp(x,n):  # 인수를 2개 받아 x의 n승을 제곱한다(3제곱 이상)
    return x**n
```

```python
# my_sq(n)에 df a열을 출력하여 나타내기
my_sq(df['a']) # 방법 1
df['a'].apply(my_sq) # 방법 2
```

```python
# a열 값에 대한 3제곱 값을 출력
df['a'].apply(my_exp,n=3) # n을 반드시 명시!
```

```python
def avg_3(x):
    # 처음에는 a열이 전달되어 a열의 0번 index 값인 10 출력
    # 다음에는 b열이 전달되어 b열의 0번 index 값인 100 출력
    return (x[0]+x[1]+x[2])/3
#     return (x[0]+x[1])/2
```
