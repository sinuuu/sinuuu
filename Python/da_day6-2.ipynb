각 컬럼별로 누락값의 갯수를 반환하는 함수

```python
import numpy as np
import pandas as pd
def count_missing(vec):
    null_vec=pd.isnull(vec) # True(NaN) or False(NaN이 아닌 값)
    null_count=np.sum(null_vec)  # True or False
    return null_count

#     print(vec)
#     print("="*50)
#     return 각 컬럼 별 누락값의 개수


cmis_col=titanic.apply(count_missing)
print(cmis_col)
#axis를 생략했으므로 기본값인 0(열 방향)
```

```python
x=[1,2,3,4,5,np.NAN]
pd.isnull(x)  # array([False, False, False, False, False,  True])
np.sum(pd.isnull(x))
# False 0, True 1로 봄
```

```python
# 열 방향으로 Null갯수와 % 알아보기
```

```python
def prop_missing(vec):
    num=count_missing(vec) # 각 컬럼에 대한 Nan의 개수
    return num/vec.size # 데이터가 몇 개 있는지
#     return 각 컬럼별 NaN의 비율


# 누락값이 아닌 것에 대한 함수 만들기
def prop_complete(vec): # Null이 아닌 값들의 비율을 리턴해주는 함수(=실제 값)
    return 1-prop_missing(vec)
```

```python
pmis_col=titanic.apply(prop_complete)
print(pmis_col)
```

```python
cmis_row=titanic.apply(count_missing,axis=1)
pmis_row=titanic.apply(prop_missing,axis=1)
pcom_row=titanic.apply(prop_complete,axis=1)

cmis_row
pcom_row # 각 행별로 누락값이 얼마나 차지하는지 알 수 있다
pmis_row
pcom_row

# titanic.loc[5] # 5는 인덱스 번호(5 인덱스에 있는 NaN 개수)
# 각 행 단위로 NaN이 몇 개 있는지 알 수 있다!


# 각 행 단위로 누락값의 개수('num_missing')을 구하여 타이타닉 데이터프레임에 추가
titanic['num_missing']=titanic.apply(count_missing,axis=1)
titanic
```

```python
# 표준점수 구하기 (어려움,...)
```

```python
def my_zscore(x): # X를 하나의 컬럼으로 생각하면 됨
    return(x-x.mean()) / x.std() # mean : 6.2. std :3.7
# x라는 표준편차를 나눈다. 분자부분은 반드시 괄호로 묶어줄 것!
# mean이나 std는 리스트 구조에서 사용할 수 없다(mean함수는 array는 적용되지만 list에는 사용불가)
# mean이나 std 함수는 넘파이에서 제공되는 함수
# 넘파이에서는 기본적으로 함수 적용 대상이 array(배열 자료형)

data=[1,3,7,9,11]
data=np.array(data)  # list 자료구조가 array로 바꿔준다는 의미(파이썬 list -> 넘파이 array로 변경됨)
my_zscore(data) 
```
