```python
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# seaborn 히스토그램 : subplots(기본틀), distplot(데이터 전달) 매서드를 사용하여 시각화

import seaborn as sns
tips=sns.load_dataset("tips")

ax=plt.subplots()
# ax=sns.distplot(tips['total_bill'],hist=False) # kde=False, hist=False 둘 중 하나 사용 : 밀집도 그래프는 제외시킨다
# 히스토그램과 밀집도 그래프가 함께 출력(넓이가 1인 그래프)


ax=sns.distplot(tips['total_bill'], rug=True) #rug(양탄자)그래프 : 선이 진할 수록 많이 몰려있다는 의미
```

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips)

#title 주기
ax.set_title("Count of Days")
```

산점도 그래프와 함께 회귀선 출력하기

```python
ax=sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
# 기본적인 산점도 그래프에 회귀선 추가
# fit_reg=False : 회계선 제거
```

```python
# jointplot 함수 : 산점도 그래프 + 히스토그램
ax=sns.jointplot(x='total_bill', y='tip',data=tips,kind='hex')

# +- 에 가까울 수록 상관관계가 높다
# + : 양의 상관관계 : x축이 커질수록 y축도 크다
# - : 음의 상관관계 : x축이 작을수록 y축도 작다
# 원 형태는 무상관


# 점이 겹쳐져 있을 때는 구분이 어렵다는 단점이 있다 -> 육각형으로 만들어주면 구분이 용이해진다(kind='hex' 사용!)
```

2차원으로 밀집도 그려보기

```python
ax=sns.kdeplot(data=tips['total_bill'], data2=tips['tip'],shade="True")
```

바 그래프 그려보기
시간(Lunch, Dinner)에 따라 지불한 비용의 평균을 바 그래프로 출력

```python
sns.barplot(x='time',y='total_bill', data=tips)
#내부적으로 평균이 구해져서 나옴 -> Dinner가 total_bill이 높게 나오는 것을 볼 수 있음

ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

```python
sns.boxplot(x='time',y='total_bill', data=tips)  #boxplot으로 바꿔주면 다양한 통계량을 한꺼번에 볼 수 있다
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

```python
# 바이올린 그래프 출력
sns.violinplot(x='time',y='total_bill', data=tips)  #boxplot으로 바꿔주면 다양한 통계량을 한꺼번에 볼 수 있다
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

관계 그래프: parirplot 매서드

```python
fig=sns.pairplot(tips) # 총 9개의 axes로 출력

# 수치데이터가 3개(total_bill,tip,size)라서 3*3으로 출력됨
# 단점 : 대각선을 기준으로 양쪽이 대칭을 이루고 있음 -> 사실상 중복된 정보가 나타난다
# x축은 독립변수, y축은 (tip)에 의해서 결과를 나타냄으로 종속변수
```

```python
#대각 요소만 작성
pair_grid=sns.PairGrid(tips) # 그리드가 만들어짐
pair_grid.map_upper(sns.regplot)
pair_grid.map_lower(sns.kdeplot) #밀집도
pair_grid.map_diag(sns.distplot,rug="True") #대각요소

# 다양한 방법으로 밀집도를 확인할 수 있다!!!
```

```python
#인덱스에 대한 조작 작업
df1=pd.DataFrame(
         np.vstack([list("abcde"), # 데이터 요소가 abcde 로 구성 
         np.round(np.random.rand(3,5),2)]).T, # T: transport / 난수가 3행 5열, 0~1의 난수 / 소수점 둘째 자리까지 출력하라는 의미
         columns=["c1","c2","c3","c4"])
df1

# np.round(3.149,2) : 소수점 둘째자리 까지 출력하라는 의미
# T : transport 사용하면 행과 열이 바뀜 (3,5)-> (5,3)
# vstack : 버티컬 방향으로 쌓는다 
# abcde의 문자열이 리스트 문자열로 들어가서 abcde가 만들어진다



df2=df1.set_index("c1") # c1열을 인덱스로 변환
df2

# 굵은 문자열은 인덱스, 안에 들어가 있는 것은 데이터
# 기존에 있던 값(열 1개가)이 사라지고 5행 3열로 변환됨
```

```python
df1=pd.read_csv("data/concat_1.csv")
df2=pd.read_csv("data/concat_2.csv")
df3=pd.read_csv("data/concat_3.csv")
```

```python
# 3개 파일 합치기(데이터 연결)
# 연결 대상 데이터 프레임들을 리스트에 담아서 전달&연결


#위에서 아래로 연결, 행과 열 이름도 같으므로 행과 열 이름이 계속 유지됨
row_concat=pd.concat([df1,df2,df3])
row_concat
```

```python
#데이터프레임에 시리즈 연결하기

#시리즈 만들기
new_row_series=pd.Series(['n1','n2','n3','n4'])
```

```python
new_row_series
```

```python
df1
```

```python
# df1에 concat 연결
pd.concat([df1,new_row_series]) 
# 데이터프레임과 시리즈를 합치면 시리즈가 새로운 열로 추가됨
# 0이라는 인덱스가 부여(열이 새롭게 추가된 것이다 보니, 0번 열에 해당하는 값이 없기 때문에 모두 NaN으로 출력)
```

```python
# 데이터프레임에 새로운 행으로 시리즈를 연결하려면 -> 불가함
# 시리즈는 열 이름이 없기 때문에 연결이 안됨(행으로는 연결 불가)
# 데이터프레임은 열, 행 이름 둘 다 있기 때문에 가능
```

```python
# 행 1개로 구성된 데이터프레임을 생성 & 연결
new_row_df=pd.DataFrame([['n1','n2','n3','n4']],
                        columns=['A','B','C','D'])
                                             
new_row_df
```

```python
df1
```

```python
#df1와 new_row_df 합치기
pd.concat([df1,new_row_df]) #데이터프레임 + 데이터프레임
```
