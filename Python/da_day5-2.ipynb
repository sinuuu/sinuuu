누락값 변경 or 삭제

```python
# fillna() : na: NaN, 특정 값으로 Na를 대체

ebola.fillna(0) # NaN 값을 모두 0으로 바꾸겠다
ebola.fillna(0).iloc[0:10,0:5] # 10행 5열로 출력
```

```python
# 순방향/역방향으로 특정값으로 대체할 때
# ebola
ebola.fillna(method='ffill')
# ffill(forward) : 순방향(위에서 아래로)으로 이동해가면서 NaN을 만나게 되면, 직전에 있던 값으로 NaN을 대체 -> NaN이 없음
```

```python
ebola.fillna(method='bfill')
# bfill(back) : 역방향(아래에서 위로)으로 이동해가면서 NaN을 만나게 되면, 직전에 있던 값으로 NaN을 대체 -> NaN이 없음
```
