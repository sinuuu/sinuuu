```python
import numpy as np
# 정렬 : sort, axis=0(행), 1(열)
a=np.array([[4,3,5,7],
          [1,12,10,9],
           [2,5,1,7]])
np.sort(a) # 기본값 axis = 1
np.sort(a,axis=1)
np.sort(a,axis=-1) # axis를 -1로 설정하면 가장 안쪽의 차원
```

```python
np.sort(a,axis=0) # 위에서 아래로 행이 이동
```

```python
# 기술통계 관련함수 : 사분위수, 중앙값, 최대/최소. 표준표차....

# 데이터 개수 알아보기
x=np.array([5,1,3,2,4,8,7,50])
len(x)
print(np.mean(x)) #1갑np.mean(x) #10
print(np.var(x))  # var : 분산  / 모분산 : n으로 나눔, 표본분산 : n-1로 나눔
print(np.sqrt(np.var(x)))
print(np.std(x))  # std : 표준편차
print(np.min(x))
print(np.max(x))
print(np.median(x)) # median : 중앙값(4와 5의 평균값을 구함)

# 사분위수 구하기
print(np.percentile(x,0)) # 최소값
print(np.percentile(x,25)) # 1사분위수
print(np.percentile(x,50)) # 2사분위수
print(np.percentile(x,70)) # 3사분위수
print(np.percentile(x,100)) # 최대값
```

```python
# stats : 기술 통계를 한방에 출력해주는 함수
from scipy.stats import describe
describe(x)
```

```python
# 난수 생성하는 함수 비교

np.random.rand(10)
np.random.rand(2,5)
# randn : 평균:0, 표준편차:1, 표준정규분포를 따르는 난수를 생수하는 함수

np.random.randn(10)
np.random.randn(2,3)


# randint : 정수 난수가 발생하는 함수
x=np.random.randint(10,size=5) # 0~9에서 난수가 1개 발생
print(x)

x=np.random.randint(10,100,size=5) # 10~99에서 난수가 1개 발생(복원 추출이라 나왔던 수가 또 나올 수 있음)
print(x)
```

```python
# 데이터 개수 세기
# unique : 중복된 값을 제거한 리스트를 만들고 싶을 때 사용하는 함수

x=np.unique([10,10,1,1,20,20,70])
x2=np.unique(x)
print(x)
print(x2)
# 중복된 값을 제거하고 중복되지 않은 값들의 array(배열)가 출력  -> array([ 1, 10, 20])

x3=np.unique(x,return_counts=True)
x3

res1,res2=np.unique(x,return_counts=True) # 두 개의 변수로도 출력이 가능하다
res1
res2
```

```python
# 유사도 알아보기

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
# 사이킷 런 패키지의 dataset 모듈에 있는 load_digits 함수를 가져와라

# 이미지 출력
digits=load_digits()
samples=[0,10,20,30,1,11,21,31]
d=[]
# digits
# digits['data'][1].reshape(8,8) #숫자가 크면 검정색, 숫자가 작으면 옅은 회색


for i in range(8):
    d.append(digits.images[samples[i]]) # sample은 images가 저장되어있는 인덱스 번호 
plt.figure(figsize=(8,2)) # 이미지를 8번 반복해라
for i in range(8):
    plt.subplot(1,8,i+1) #1행 8열, i는 0부터 7, 들어가야할 위치를 의미)
    plt.imshow(d[i],interpolation='nearest',cmap=plt.cm.Blues) # 숫자 데이터를 이미지로 변환하여 출력
    # interpolation : 부드럽게, bone_r : 흑백출력 
    plt.title('img{}'.format(i+1)) # 일련번호
    
plt.tight_layout()    
plt.show()
    

    
# len(digits['data']) # 숫자 이미지가 1797번까지 있다

# digits['data'][10]
# digits['target_names'][digits['target'][10]]
```

```python
v=[]
np.shape(d[0]) # (8, 8)
# sample의 이미지가 8개라서 8,8


for i in range(8):
    v.append(d[i].reshape(64,1)) # 8,8 -> 64,1 로 바꾸겠다

plt.figure(figsize=(8,3)) # 이미지를 8번 반복해라
for i in range(8):
    plt.subplot(1,8,i+1) #1행 8열, i는 0부터 7, 들어가야할 위치를 의미)
    plt.imshow(v[i],interpolation='nearest',cmap=plt.cm.Oranges) # 숫자 데이터를 이미지로 변환하여 출력
    # interpolation : 부드럽게, bone_r : 흑백출력 
    plt.title('img{}'.format(i+1)) # 일련번호
    
plt.tight_layout(w_pad=10)    
plt.show()
```

```python
# 이미지에 대한 변환
# 텐서(tenser) : 같은 크기의 행렬이 여러 개 같이 묶여 있는 것 (다차원 배열)

from scipy import misc
img_rgb=misc.face()
img_rgb.shape # (768, 1024, 3) # 픽셀, 뒤에 3은 color 

plt.subplot(221)
plt.axis('off') # 축을 없앰
plt.imshow(img_rgb)

plt.subplot(222)
plt.axis('off')
plt.imshow(img_rgb[:,:,0],cmap=plt.cm.gray)  # 0: red  1: green   2: blue
# red에 해당하는 수치값만 gray값으로 나옴

plt.subplot(223)
plt.axis('off')
plt.imshow(img_rgb[:,:,1],cmap=plt.cm.Blues)

plt.subplot(224)
plt.axis('off')
plt.imshow(img_rgb[:,:,2],cmap=plt.cm.RdYlGn_r)

plt.show
```
