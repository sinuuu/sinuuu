 데이터 재구조화 연습 데이터

```python
import numpy as np
import pandas as pd
data = pd.DataFrame({'cust_id': ['c1', 'c1', 'c1', 'c2', 'c2', 'c2', 'c3', 'c3', 'c3'], 
'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'], 
'pch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]})

# 행 인덱스 별도 부여하지 않았기 때문에 0~8 임의 부여
```

```python
data
```

```python
# 행 인덱스 c1,c2,c3
# 열 인덱스 p1,p2,p3 로 바꾸고 pch_amt값을 넣겠다!

# ★ pivot 배열
data_pivot=data.pivot(index='cust_id',columns='prod_cd',values='pch_amt')
# pivot은 데이터프레임이 가지고 있는 함수
data_pivot
```

```python
# pivot_table 배열
pd.pivot_table(data,index='cust_id',columns='prod_cd',values='pch_amt')
# pivot_table은 판다스가 가지고 있는 함수
# 판다스 안에는 데이터프레임과, 피벗함수가 다 들어있다( 판다스 > 데이터프레임, 피벗T)
```

```python
# -> pivot_table을 주로 사용

# 인덱스를 2개인 경우 pivot은 에러
# data.pivot(index=['cust_id','grade'],columns='prod_cd',values='pch_amt')

# 인덱스가 2개인 경우 pivot_table은 정상 수행
pd.pivot_table(data,index=['cust_id','grade'],columns='prod_cd',values='pch_amt')

# 컬럼이 2개인 경우 pivot은 에러
# data.pivot(index='cust_id',columns=['grade','prod_cd'],values='pch_amt')

# 컬럼이 2개인 경우 pivot은 에러
pd.pivot_table(data,index='cust_id',columns=['grade','prod_cd'],values='pch_amt')


# -> 데이터에 값이 없으면 NaN으로 기입됨
```

```python
data
```

```python
# 행인덱스 A,B 열인덱스 p1,p2,p3 으로 해서 Pch_amt를 데이터로 주고 싶을 때

# data.pivot(index='grade',columns='prod_cd',values='pch_amt')
# pivot 함수는 중복값이 있는 경우 에러가 발생(인덱스가 여러 개가 있어서)


pd.pivot_table(data,index='grade',columns='prod_cd',values='pch_amt')
# 열 행 인덱스가 1개이지만 데이터 값이 여러개인 경우, 자동으로 평균값(np.mean)이 출력됨

# pd.pivot_table(data,index='grade',columns='prod_cd',values='pch_amt',aggfunc=np.count_nonzero,margins=True)
pd.pivot_table(data,index='grade',columns='prod_cd',values='pch_amt',aggfunc=np.mean,margins=True)
# -> mean하면 합계의 평균이 출력

# aggfun(집계함수) : 데이터들을 합쳐서 계산하는 함수
# mean, sum 등 적용하고자 하는 집계함수를 기입하면 됨

# margins : 열 합계와 행 합계가 다 출력됨(기본적으로 False라 True로 설정)
```

stack / unstack 함수

```python
mul_index=pd.MultiIndex.from_tuples([('cust1','2019'),('cust1','2020'),('cust2','2019') ,('cust2','2020')])
mul_index

# ['cust1', 'cust2'] : 대분류, ['2019', '2020'] : 중분류
# MultiIndex.from_tuples : 인덱스를 2개이상 겹겹히 이용하려고 할 때 사용

from pandas import DataFrame as df # DataFrame을 df로 만든다
data=DataFrame(np.arange(16).reshape(4,4),
               columns=['p1','p2','p3','p4'],
               index=mul_index) # mul_index를 인덱스로 삼고 싶다!
data

# 대분류에는 cust1, 중분류에는 2019, 소분류에는 p1~p2로 나열하고자 함 => 이때 사용하는 'stack 함수'
# cust1  2019   p1  0
#               p2  1

data_stacked=data.stack()
data_stacked
```

```python
data_stacked['cust1']['2019'][['p3','p4']]
#       -> 대분류   중분류  소분류 이렇게 분류가 가능하다 (2개 이상의 값도 가능)
```

```python
# 결측값이 있을 때 제거를 할 것인지, Nan으로 설정할지 지정하는 옵션

data

# data의 cust2에 대한 p4값을 모두 nan으로 설정(np.nan)
data.loc['cust2','p4']=np.nan
data
```

```python
# 여러 개의 열을 하나로 정리하기
weather=pd.read_csv("data/weather.csv")
weather

# 날짜 열을 행 데이터로 피벗
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],var_name='day',value_name='temp')
weather_melt

# 'id','year','month','element' 가 행 인덱스로, 열에는 tmax, tmin으로 값이 오게 작성
weather_tidy=weather_melt.pivot_table(index=['id','year','month','day'],
                                      columns='element',
                                      values='temp')
weather_tidy


weather_tidy_flat=weather_tidy.reset_index()
weather_tidy_flat
```
