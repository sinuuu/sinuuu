concat : 데이터프레임을 합치는 과정

```python
import pandas as pd
df1=pd.read_csv("data/concat_1.csv")
df2=pd.read_csv("data/concat_2.csv")
df3=pd.read_csv("data/concat_3.csv")
df3
```

```python
# 데이터프레임 연결(위에서 아래 방향으로)
row_concat=pd.concat([df1,df2,df3])
```

```python
row_concat
# 같은 값이라도 행 인덱스가 다름
```

```python
#a7, b7, c7, d7 을 출력하려면,
row_concat.iloc[7]
```

```python
nrs=pd.Series(['s1','s2','s3','s4'])  #행이 아니라 열로 추가됨
nrs
```

```python
pd.concat([df1, nrs]) # 열이 만들어지면서 NaN값이 추가됨 
```

```python
# ignore_index=True : 기존의 인덱스를 버리고 행 인덱스를 다시 부여함
row_concat_i=pd.concat([df1,df2,df3],ignore_index=True)
row_concat_i
```

```python
# 열 방향으로 합치고 싶을 때
# pd.concat([df1,df2,df3],axis=0) # 행 방향, 기본값 : 0
col_concat=pd.concat([df1,df2,df3],axis=1) # 열 방향, 기본값 :1

# a라는 열 데이터를 추출할 때
col_concat['A']
```

```python
col_concat=pd.concat([df1,df2,df3],axis=1,ignore_index=True)
col_concat
```

```python
# col_concat에 새로운 열을 추가할 때
col_concat['new_col_list']=['n1','n2','n3','n4'] # [추가할 인덱스]=[추가할 데이터]
col_concat
```

```python
# 열 이름의 일부가 서로 다른 데이터프레임 연결

df2.columns=['E','F','G','H']
#['A', 'B', 'C', 'D']
df2
```

```python
df3.columns=['A','C','F','H']
df3
```

```python
row_concat=pd.concat([df1,df2,df3])
row_concat

# 합치기를 하면 데이터프레임에 데이터가 없으면 NaN으로 표기됨
```

```python
# 데이터프레임에 NaN값 없이 연결해보기(공통된 열만 골라서 연결 : join속성 -> inner로 지정)
```

```python
pd.concat([df1,df2,df3],join='inner')
# df1, df2, df3이 공통된 열이 업으므로 아무것도 출력되지 않음
```

```python
# df1, df3의 공통된 열 연결
pd.concat([df1,df3],join='inner',ignore_index=True) #df1, df3 공통된 열 연결
```

```python
# 데이터 프레임을 행 방향으로 연결하기
df1.index #0~3까지의 인덱스를 가지고 있다는 뜻
df2.index=[4,5,6,7]
```

```python
df3.index=[0,2,5,7]
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
# df1과 df3의 공통 행만 연결    -> #0~3만 연결 예상
pd.concat([df1,df3],axis=1,join='inner') # 행으로 연결하기 위해 axis값은 반드시 1
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])

df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
# merge: 공통된 열을 기준으로 병합(고객번호), 양쪽 모두 키가 존재하는 데이터만 연결(내부조인, innter join)
# 외부조인(outer join) : 키 값이 한쪽에만 있어도 데이터를 병합
pd.merge(df1,df2,how='inner') #디폴트 :inner 조인
pd.merge(df1,df2,how='outer') # outer 조인

# 키가 없는 값은 NaN 값으로 나옴
```

```python
# 외부조인은 2가지 종류가 있다

# outer join은 left/right outer join으로 구분
pd.merge(df1,df2,how='left')
# outer join보다 하나가 덜 나옴 -> left : 기준 데이터 프레임(왼쪽에 있는 데이터프레임을 기준으로 삼겠다는 의미)
# -> 왼쪽 데이터프레임을 모두 포함해서 연결함(왼쪽에는 1008번이 없기 때문에 출력이 안됨)
```

```python
pd.merge(df1,df2,how='right') #right : 기준 데이터 프레임(오른쪽에 있는 데이터프레임을 기준으로 삼겠다는 의미)
print(df2)
```
