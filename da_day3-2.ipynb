```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# anscome[anscombe['dataset']==I]
dataset_1=anscombe[anscombe['dataset']=='I'] #데이터 그룹 내 I을 추출하여 저장
dataset_2=anscombe[anscombe['dataset']=='II']
dataset_3=anscombe[anscombe['dataset']=='III']
dataset_4=anscombe[anscombe['dataset']=='IV']
dataset_4
```

```python
#선 그래프 : plot 매서드
# plot 매서드에 x, y축 데이터를 전달 -> 선 그래프 출력
plt.plot(dataset_1['x'], dataset_1['y'],'o')

#'o'옵션을 주면 점으로 출력된다
```

```python
# plt.plot(dataset_1['x'], dataset_1['y'],'o')
#데이터를 나타내는 부분이 점으로 출력된다
```

```python
# 시각화 과정
# 1) 전체 그래프가 출력될 기본 틀(figure) 생성
fig=plt.figure()

# 2) 틀을 그래프 격자로 나눔(그래프 격자 생성)
axes1=fig.add_subplot(2,2,1) # 틀을 2*2로 나누고 1번을 위치(왼쪽 상단부터 시작)을 axas1
axes2=fig.add_subplot(2,2,2) 
axes3=fig.add_subplot(2,2,3) 
axes4=fig.add_subplot(2,2,4)

# 3) 각 격자에 그래프를 추가
axes1.plot(dataset_1['x'], dataset_1['y'],'o')
axes2.plot(dataset_2['x'], dataset_2['y'],'o')
axes3.plot(dataset_3['x'], dataset_3['y'],'o')
axes4.plot(dataset_4['x'], dataset_4['y'],'o')

# 4) 각 격자에 그래프 출력
# 타이틀 출력
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig.suptitle("Anscombe Data")
fig.tight_layout() # 간격을 줄 수 있다
```

```python
# 시각화도구: matplotlib, seaborn
tips=sns.load_dataset("tips")
tips
print(tips['total_bill'].max()) #->최대값
print(tips['total_bill'].min()) #->최소값

#팁을 준 사람들의 분석표
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1) #격자 추가 1칸만 만들겠음 / axis(축) axes(축들, 격자 1개)

#axes 설정(x,y 축 설정)
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
axes1.set_title("Histogram of Total Bill")  # 타이틀 주기

#히스토그램 추가
# axes1.hist(tips['total_bill'])  #-> y축 값
axes1.hist(tips['total_bill'],bins=20) #-> 구간 나누기(20개로)
```

```python
# 산점도 그래프(skatter plot) : 변수 2개로 그린 그래프 , 이변량 그래프
# 식사 비용에 따른 팁 열의 분포를 산점도 그래프로 나타내보자
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
axes1.set_title("Scatterplot of Total Bill vs Tip") 

# 이상치(outlier)를 찾을 때 산점도 그래프로 나타내면 좋다!(box plot으로도 가능)
```

box plot

```python
boxplot=plt.figure()
axas1=boxplot.add_subplot(1,1,1)

# tips=[tips['sex']=='Female']  #-> 성별이 Female인 사람의 데이터만 추출됨
# x축은 성별을 나타내는 이산형 변수, y축은 팁을 나타내는 연속형 변수 출력

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
               labels=['Female','Male'])

#boxplot([   ,    ],labels)
```

 다변량 그래프 그리기 : 변수가 3개

```python
def recode_sex(sex):
    if sex=="Female":
        return 0
    else:
        return 1
```

```python
tips['sex_colortips['sex'].apply(recode_sex)
# 작성형식 : 데이터.apply(적용함수)  => Female 0 /Male 1 로 변경
# apply : 특정 데이터에 대해 특정 함수를 적용하고자 할 때 사용
```
