```python
# 100만개 데이터 셋
# 6000명 4000편 100만개 영화평점
# 평점, 사용자정보, 영화정보


import pandas as pd
unames=['user_id', 'gender', 'age', 'occupation', 'zip']
rnames=['user_id', 'movie_id', 'rating', 'timestamp']
mnames=['movie_id', 'title', 'genres']

users=pd.read_csv("data/movielens/users.dat", names=unames, sep="::", header=None)
print(users.head(5)) # Gender::Age::Occupation::Zip-code

ratings=pd.read_csv("data/movielens/ratings.dat", names=rnames, sep="::", header=None)
print(ratings.head(5)) # UserID::MovieID::Rating::Timestamp

movies=pd.read_csv("data/movielens/movies.dat", names=mnames, sep="::", header=None)
print(movies.head(5)) # MovieID::Title::Genres
```

```python
# users.columns=['user_id', 'gender', 'age', 'occupation', 'zip']
```

```python
# 데이터 합칠 때는 merge, cincat 함수
# pd.merge(왼쪽, 오른쪽, on=기준변수, left_on, right_on... )
pd.merge(ratings, users)
#pd.merge(ratings, users, on='user_id') 겹치는 게 user_id 하나라 기준으로 주지 않아도 알아서 공통을 찾아 병합할 수 있음.
```

```python
data=pd.merge(pd.merge(ratings, users),movies) # merge 두 번 사용 가능.
data.info()
#data의 0번행 추출 iloc(정수위치), loc(라벨),ix
data.loc[0]

#성별에 따른 각 영화의 평균 평점
mean_ratings=data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings.head(5)
```

```python
import nltk
nltk.download('punkt')
# 자연어처리: 데이터를 (토큰화->정제->정규화)
# 토큰화 : 문장을 단어로 나누는 작업 (How are you?) -> how, are, you
# 정제: 특수문자, 구두점 등 모두 제거하는 작업
# ex) Don't Don t-> do not 으로 될 수 있어서 정제할 때 조심히 해야 함
from nltk.tokenize import word_tokenize
print(word_tokenize("Don't be fooled by the dark sounding name, Mr.Jone's"))
```

```python
# 영화제목으로 그룹화-> size로 제목별 평점 정보 건수를 추출
ratings_by_title=data.groupby('title').size()
ratings_by_title[:10]

# 250건 이상의 평점 정보가 있는 영화만 추출 불린추출로
ratings_by_title[ratings_by_title>=250]
len(ratings_by_title[ratings_by_title>=250]) # 총 1216편  ratings_by_title[ratings_by_title>=250].size 도 됨

# 250건 이상의 평점 정보가 있는 영화제목만 추출
active_titles=ratings_by_title.index[ratings_by_title>=250]
active_titles

# mean_ratings에서 acrive_titles 추출
mean_ratings.loc[active_titles] # mean_ratings.ix[active_titles]와 같음

# 여성에게 높은 평점을 받은 영화 목록을 출력(상위10개)
top_female_ratings=mean_ratings.sort_index(by="F", ascending=False) # ascending=False 내림차순
top_female_ratings[:10]
```

```python
# 남녀간 평균 평점의 차이가 가장 큰 영화가 무엇인지.(남)
mean_ratings['diff']=mean_ratings["M"]-mean_ratings["F"]
sorted_by_diff=mean_ratings.sort_values(by="diff")
sorted_by_diff
```

```python
rating_std_by_title=data.groupby('title')['rating'].std() #영화 제목별 rating에 대한 표준편차
rating_std_by_title.sort_values(ascending=False)[:10]
```
